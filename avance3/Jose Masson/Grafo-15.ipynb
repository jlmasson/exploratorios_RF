# Degree Distribution (15 minutos)

Para esta parte se proceder√° a trabajar con el top 3 donde existen la mayor cantidad de conexiones
* Aeropuerto
* Mall del Sur
* City Mall

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Aeropuerto

```python
aeropuerto = open("./15/aeropuertoPeso2.csv", 'rb')
aeropuerto.readline()
conexiones_aeropuerto = nx.read_edgelist(aeropuerto, delimiter=',', nodetype=int, data=(('weight',int),))
aeropuerto.close()
```

```python
degrees_aeropuerto = conexiones_aeropuerto.degree()

degrees = list(degrees_aeropuerto.values())

degrees = sorted(set(degrees))
histogram = [list(degrees_aeropuerto.values()).count(x) for x in degrees]
```

```python
plt.figure()
plt.plot(degrees, histogram)
plt.xlabel('Degree')
plt.ylabel('Number of nodes')
plt.title('Aeropuerto')
plt.savefig("aeropuerto_15.png", dpi=500)
plt.show()
```

```python
plt.figure()
plt.loglog(degrees,histogram, basex=10, basey=10) # Log Log
plt.xlabel('Degree')
plt.ylabel('Number of nodes')
plt.title('Aeropuerto')
plt.savefig("aeropuerto_15_pl.png", dpi=500)
plt.show()
```

```python
items = list(degrees_aeropuerto.items())
items.sort(key=lambda x: x[1], reverse=True)

minimum_degree = items[-1][1]
maximum_degree = items[0][1]
```

```python
print("Minimum degree: {:.2f}".format(minimum_degree)) 
print("Maximum degree: {:.2f}".format(maximum_degree))
print("Node ID with maximum degree: {}".format(items[0][0]))
print("Node ID with minimum degree: {}".format(items[-1][0]))
```
